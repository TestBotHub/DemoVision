```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import cv2
from sklearn.externals import joblib
from sklearn import datasets
from skimage.feature import hog
from sklearn.svm import LinearSVC
from scipy import ndimage
```

```python
camera = cv2.VideoCapture(1)
# while True:
retval, im = camera.read()
im = im[147:443,68:530]
im = ndimage.rotate(im, 90, reshape = True)
cv2.imwrite("test.png", im)
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5,5), 0)
ret, th = cv2.threshold(blur, 100, 300, cv2.THRESH_BINARY_INV)
plt.imshow(th)
#gray2 = cv2.bilateralFilter(gray, 11, 17, 17)
#edged = cv2.Canny(gray2,30, 40)
#plt.imshow(edged, cmap='gray')
cv2.waitKey(10)
```

```python
_, ctrs, hier = cv2.findContours(th.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
ctrs
```

```python
rects = [cv2.boundingRect(ctr) for ctr in ctrs]
rects = sorted(rects, key=lambda x:x[0])
imgs = []
im2 = im.copy()
for rect in rects:
    if rect[2] * rect[3] > 50 and rect[2]*rect[3] < 200:
        cv2.rectangle(im, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (0, 255, 0), 3)
        imgs.append(im2[rect[1]:rect[1]+rect[3], rect[0]:rect[0]+rect[2]])
        print(rect)
plt.imshow(im)
```
